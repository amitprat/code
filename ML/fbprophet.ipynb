```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
import datetime as dt;

df = pd.read_csv('C:\\Users\\amisi\\Downloads\\stock_data\\all_stocks_5yr.csv')
df.head()
```

```python
df = df.rename(columns={'Name': 'Ticks'})
df.head()
```

```python
amzn = df.loc[df['Ticks'] == 'AMZN']
amzn
```

```python
amzn_df = amzn.copy()

# Change to datetime datatype.
amzn_df.loc[:, 'date'] = pd.to_datetime(amzn.loc[:,'date'], format="%Y-%m-%d")
amzn_df
```

```python
# First Subplot
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,5))
ax1.plot(amzn_df["date"], amzn_df["close"], color="red")
ax1.set_xlabel("Date", fontsize=12)
ax1.set_ylabel("Stock Price")
ax1.set_title("Amazon Close Price History")

# Second subplot
ax1.plot(amzn_df["date"], amzn_df["high"], color="green")
ax1.set_xlabel("Date",fontsize=12)
ax1.set_ylabel("Stock Price")
ax1.set_title("Amazon Close High Price")

# Third subplot
ax1.plot(amzn_df["date"], amzn_df["low"], color="yellow")
ax1.set_xlabel("Date",fontsize=12)
ax1.set_ylabel("Stock Price")
ax1.set_title("Amazon Close Low Price")

# Fourth subplot
ax2.plot(amzn_df["date"], amzn_df["volume"], color="orange")
ax2.set_xlabel("Date", fontsize=12)
ax2.set_ylabel("Stock Volume")
ax2.set_title("Amazon Stock Volume History")

plt.show()
```
